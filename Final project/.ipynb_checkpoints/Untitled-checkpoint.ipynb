```python
# 
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
MSE_average_KFOLDS=0
print('Linear regresion with 20 KFlod')
from sklearn.cross_validation import KFold
kf = KFold(df_binary.shape[0], n_folds=20,shuffle=True)
x=df_binary.iloc[0:1161,0:62]
y=df_binary.iloc[0:1161,62:63]
i=1
for train_index, test_index in kf:
    trainx= x.iloc[train_index]
    testx = x.iloc[test_index]
    trainy =y.iloc[train_index]
    testy =y.iloc[test_index]
# Create linear regression object
    regr = linear_model.LinearRegression(normalize =True)

# Train the model using the training sets
    regr.fit(trainx, trainy)

# Make predictions using the testing set
    prediction = regr.predict(testx)
    MSE_average_KFOLDS+=mean_squared_error(testy, prediction)
    #print('Score for Kfold %d : %f'%(i,regr.score(testx,testy)))
    print('MSE for Kfold %d : %f'%(i,mean_squared_error(testy, prediction)))
    i+=1

MSE_average_KFOLDS=MSE_average_KFOLDS/20
print('MSE_average_KFOLDS :',MSE_average_KFOLDS)


```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.cross_validation import KFold
kf = KFold(df_binary.shape[0], n_folds=20,shuffle=True)
x=df_binary.iloc[0:1161,0:62]
y=df_binary.iloc[0:1161,62:63]
MSE_average_KFOLDS=0

for train_index, test_index in kf:
    trainx= x.iloc[train_index]
    testx = x.iloc[test_index]
    trainy =y.iloc[train_index]
    testy =y.iloc[test_index]
    regressor = DecisionTreeRegressor()
    regressor.fit(trainx, trainy)
    prediction=regressor.predict(testx)
    MSE_average_KFOLDS+=mean_squared_error(testy, prediction)
    print('Scoreeee',regr.score(testx,testy))

MSE_average_KFOLDS=MSE_average_KFOLDS/20
print('MSE_average_KFOLDS :',MSE_average_KFOLDS)

```
