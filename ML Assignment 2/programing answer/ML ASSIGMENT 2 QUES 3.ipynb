```python
import csv
import re  
from sklearn.model_selection import KFold
# reading csv file 
#with open('reviewstrain.txt', 'r') as csvfile: 
    # creating a csv reader object 
 #   csvreader = (csv.reader(csvfile)) 
    
  #  for row in csvreader:
   #     print(row)
        #x,y=str(str(row).strip()).split(" ",1)
        #//print(x)
        #//print(y)
        
trainrow=[]        
file = open('reviewstrain.txt', 'r')  
lines=file.readlines()

for line in lines:
    line=line.strip(('\t'))
    line=line.strip(('\n'))
    x,y=line.split(" ",1)
    trainrow.append((x,re.sub("\t|\n","",y)))
 #   print(x)
#    print(y)

#print(row)
print(len(trainrow))
Num_of_example=len(trainrow)
```

```python

def takeSecond(elem):
    return elem[1]
accuracy1=0
kf = KFold(n_splits=5, shuffle=False)
for train_indices, test_indices in kf.split(trainrow):
    #print('Train: %s | test: %s' % (train_indices, test_indices))
    KKK=3

    prediction5=[]
    for i in test_indices:
        distancelist=[]
        temp1=[]
        y=trainrow[i]
        for i2 in (y[1]).split(" "):
            temp1.append(i2) # All tokens for a comment
        temp1=list(set(temp1)) # unique tokens for a comment
        for i3 in train_indices:
            temp2=[]
            y2=trainrow[i3]
            for i4 in (y2[1]).split(" "):
                temp2.append(i4)# All tokens for a comment
            temp2=list(set(temp2))    # unique tokens for a comment
            distance=0 # difference between two comments
            for val in temp1 :
                if(val in temp2):
                    distance+=1
            if(distance==0):
                distance=1000000000
            else:
                distance=(1/distance) # distance according to the formula
            distancelist.append((int(y2[0]),distance))    #labels with distance 
        #sorted(distancelist,key=distancelist[1])
        distancelist.sort(key=takeSecond) # sorted according to distance
        #print(distancelist)
        class0=0
        class1=0
        #print((distancelist[0])[1])
        mindis=(distancelist[0][1])
       # print(type((distancelist[0])[0]))
       # print(type((distancelist[0])[1]))
        if(((distancelist[0])[0])==1):
            class1+=1
        else:
            class0+=1
        # CODE FOR ADDING THE PREDICTION

        k=2
        #for i23 in range(1,25)
      #  print(i)
       # print(distancelist[0:50])
        for i5 in range(1,len(distancelist)):
          #  print("new min",mindis,'----',(distancelist[i5])[1])
            if k<=KKK or mindis==(distancelist[i5])[1]:
             #   print("k=",k)
              #  print(mindis,"-----",(distancelist[i5])[1])
                if((distancelist[i5])[0]==1):
                    class1+=1
                    mindis=(distancelist[i5])[1]
                    k=k+1
                else:
                    class0+=1
                    mindis=(distancelist[i5])[1]
                    k=k+1
            else :
                break
        if(class1>=class0):
            prediction5.append(1)
        else:
            prediction5.append(0)
    FP=0
    FN=0
    TP=0
    TN=0
    jj=0
    for i in test_indices:
        if(int((trainrow[i])[0])==1 and prediction5[jj]==1):
            TP+=1
            jj+=1
        elif(int((trainrow[i])[0])==1 and prediction5[jj]==0):
            FN+=1
            jj+=1
        elif(int((trainrow[i])[0])==0 and prediction5[jj]==0):
            TN+=1
            jj+=1
        elif(int((trainrow[i])[0])==0 and prediction5[jj]==1):
            FP+=1
            jj+=1
  #  print("TP= ",TP)
   # print("FP= ",FP)
   # print("TN= ",TN)
   # print("FN= ",FN)
    accuracy1+=(TP+TN)/(TP+TN+FP+FN)
   # print("Accuracy =",accuracy)
    
accuracy1=accuracy1/5    
print("accuracy for k=3 =====",accuracy1)
        
```
